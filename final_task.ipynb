При p-value, равном 0.00002, меньшем, чем уровень значимости alpha = 5%, мы можем отвергнуть нулевую гипотезу о нормальности распределения метрики CPA в выборке.

Для проверки первоначальной гипотезы о равенстве средних значений метрики CPA в двух группах можно использовать t-тест Стьюдента для независимых выборок. Однако, в данном случае мы отвергли гипотезу о нормальности распределения, поэтому использование t-теста может быть неправильным. В этом случае лучше использовать непараметрический критерий Манна-Уитни, который не требует предположения о нормальности распределения.
