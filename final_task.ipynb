4) Вы решили сравнивать метрику CPA в двух группах. Размер выборки - 2350 элементов в каждой группе. Для проверки нормальности распределения на выборке в 2350 наблюдений применили , критерий Шапиро-Уилка и получили p-value, равный 0.00002, alpha = 5%. Какой бы вывод мы могли сделать в данном случае? В этом случае какой статистический критерий для проверки первоначальной гипотезы тут лучше всего подойдёт и почему ?

При p-value, равном 0.00002, меньшем, чем уровень значимости alpha = 5%, мы можем отвергнуть нулевую гипотезу о нормальности распределения метрики CPA в выборке.

Для проверки первоначальной гипотезы о равенстве средних значений метрики CPA в двух группах можно использовать t-тест Стьюдента для независимых выборок. Этот критерий подходит для проверки различий между двумя выборками при условии, что распределение метрики в каждой из групп близко к нормальному.

Однако, в данном случае мы отвергли гипотезу о нормальности распределения, поэтому использование t-теста может быть неправильным. В этом случае лучше использовать непараметрический критерий Манна-Уитни, который не требует предположения о нормальности распределения.

5)  Мы провели АБ-тест на увеличение average timespent  per user. По итогам тестирования мы получили следующие данные. Является ли результат статистически значимым с уровнем доверия 80%? Какую версию мы выкатим на продакшн? A) Средняя - 360, отклонение - 40, количество - 9802 B) Средняя - 352, отклонение - 58, количество - 9789

```python
import math
mean_a = 360
std_a = 40
n_a = 9802
mean_b = 352
std_b = 58
n_b = 9789
t_crit = 1.282
```

```python
se_a = std_a / math.sqrt(n_a)
se_b = std_b / math.sqrt(n_b)
t_stat = (mean_a - mean_b) / math.sqrt(se_a**2 + se_b**2)
print("t-статистика:", t_stat)

if abs(t_stat) > t_crit:
    print("Различия между группами статистически значимы")
    if mean_a > mean_b:
        print("Выбираем версию A")
    else:
        print("Выбираем версию B")
else:
    print("Различия между группами не статистически значимы")
```
